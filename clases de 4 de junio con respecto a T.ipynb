```python
#SIMULADOR DE LA ECUACION DE DIFUSIVIDAD
import numpy as np
```

```python
P=[0,4000,4000,4000,0]
Nodos=len(P)
Nodos2=[0,1,2,3,4]
Tiempo=1000
dt=10
dx=200
#condi iniciales
t=dt
Pin=P
Pin2=Pin #resultado
Pdt=[0,0,0,0,0,0]
```

```python
def Predt(P1,P2,P3,dt,dx):
  return(((P1-2*P2+P3)/(dx*dx))*dt)+P2
```

```python
Predt(P[0],P[1],P[2],dt,dx)
```

```python
while t<Tiempo:
  for i in (Nodos2):
    if i==0:
      Pdt[i]=P[0]
    elif i==len(Nodos2)-1:
      Pdt[i]=P[len(Nodos2)-1]
    else:
     Pdt[i]=Predt(Pin[i-1],Pin[i],Pin[i+1],dt,dx)
   
  t=t+dt
  Pdt=Pin
  Pin2=np.append(Pin2,Pdt)
  
```

```python

print(Pin2)
```
